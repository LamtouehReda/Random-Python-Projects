```python
import numpy as np
from scipy.optimize import minimize

def objective(x):
    return (x[0]+x[1])
def constraint(x):
    return 2-(x[0]**2+x[1]**2)
x=[0,0]
bnds=[(-4,4),(-4,4)]
#it could be more than 1 constraint
cons=[{'type':'eq','fun':constraint}]
sol=minimize(objective,x,method='SLSQP',bounds=bnds,constraints=cons)
```
